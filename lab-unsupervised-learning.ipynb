```python
# K-means
# visualizing clusters
from sklearn import cluster, datasets

n_samples = 1500
X, _ = datasets.make_blobs(n_samples=n_samples, centers=5, cluster_std=0.7, n_features=2, random_state=0)
# X, y = datasets.make_blobs(n_samples=n_samples, centers=3, cluster_std=0.7, n_features=2, random_state=0)
# features are x1 and x2.

plt.figure(figsize=(5, 5))
plt.scatter(X[:, 0], X[:, 1], alpha=0.5,) # c=y colors
# plt.scatter(X[:, 0], X[:, 1], c=y, alpha=0.5,) # c=y colors

plt.show()
```

```python
# visualizing clusters

# %%time

kmeans = cluster.KMeans(n_clusters=5) # look for 5 clusters
kmeans.fit(X)
pred = kmeans.predict(X) # give me value for each points
# cluster: group of points that have centroids

plt.figure(figsize=(5, 5))
plt.scatter(X[:, 0], X[:, 1], c=pred,alpha=0.5)
plt.show()

# describe difference: borders are created
```
